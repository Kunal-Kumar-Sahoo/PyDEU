```python
principle = 10_000
interest_rate = 0.05
years = 20
values = []

for i in range(years + 1):
    values.append(principle)
    principle += principle * interest_rate
pylab.plot(values)

pylab.title('{}% Growth, Compounded Annually'.format(interest_rate * 100))
pylab.xlabel('Years of compounding')
pylab.ylabel('Value of principle (₹)')
```

```python
pylab.plot(values, 'ko')
```

```python
pylab.plot(values, linewidth=30)
pylab.title('{}% Growth, Compounded Annually'.format(interest_rate * 100), fontsize='xx-large')
```

### Practice problem:
### Draw two different plots in the same graph
    * Pie chart
    * Line graph

```python
languages = ['Python', 'C++', 'Java', 'Javascript', 'Dart', 'Golang', 'Julia', 'Kotlin']
population = [randrange(0, 100) for _ in range(len(languages))]

plt.subplot(1, 2, 1)
explosion = [0 for _ in languages]
explosion[languages.index('Python')] = 0.15
# print(explosion)

plt.pie(population, labels=languages, explode=explosion)

plt.subplot(1, 2, 2)
plt.plot(languages, population)
plt.xticks(rotation=45)

plt.show()
```
